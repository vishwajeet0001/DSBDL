```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Iris.csv')
```

```python
df
```

```python
#1
print("Features and their types:")
print(df.dtypes)
```

```python
#2.Create a histogram for each feature
plt.figure(figsize=(10, 6))
for feature in df.columns[:-1]:
    sns.histplot(df[feature], kde=True, label=feature, alpha=0.7)
plt.title('Histograms of Features in Iris Dataset')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
#3 Create a box plot for each feature
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop('Species', axis=1))
plt.title('Box Plots of Features in Iris Dataset')
plt.xlabel('Feature')
plt.ylabel('Value')
plt.show()
```
